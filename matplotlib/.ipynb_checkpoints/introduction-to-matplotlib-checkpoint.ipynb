```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Introduction to Matplotlib

```python
plt.plot([1, 2, 3, 4, 5])
```

```python
x = [1, 2, 3, 4, 5]
y = [11, 22, 33, 44, 55]
plt.plot(x, y)
```

```python
# 1st Method
fig = plt.figure() #create a figure
ax = fig.add_subplot() # adds axes
plt.show()
```

```python
# 2nd Method
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x, y)
```

# Matplotlib Example

```python
# Prepare Data
x = [1, 2, 3, 4, 5, 6]
y = [1, 4, 9, 16, 25, 36]

# Setup Plot
fig, ax = plt.subplots(figsize=(10, 10))

# Plot Data
ax.plot(x, y)

# Customize Plot
ax.set(title="Square Plot",
       xlabel="x-axis",
       ylabel="y-axis")

fig.savefig("square-plot.png")
```

# Figures using Numpy

```python
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Line Plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Scatter Plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.cos(x))
```

```python
# Bar Plot
butter_prices = { "Almond Butter": 10, "Peanut Butter": 8, "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(butter_prices.keys(), butter_prices.values())
ax.set(title="Butter Price Catalog", ylabel="Price ($)"); # ";" removes the object info line
```

```python
# Horizontal Bar Graph
fig, ax = plt.subplots()
ax.barh(list(butter_prices.keys()), list(butter_prices.values()));
```

```python
# Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Option 1: Subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
x = np.linspace(0, 10, 100)
ax1.plot(x, x/2)
ax2.plot(x, x**2)
ax3.barh(list(butter_prices.keys()), list(butter_prices.values()))
ax4.plot(x, np.sin(x))
```

```python
# Option 2: Subplots
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
ax[0, 0].plot(x, x**2)
ax[0, 1].plot(x, x/2)
ax[1, 0].barh(list(butter_prices.keys()), list(butter_prices.values()))
ax[1, 1].plot(x, np.sin(x));
```
